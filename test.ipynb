```python
import tkinter as tk
from tkinter import font, messagebox

class ChessboardApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Interactive FEN Chessboard")
        self.root.geometry("800x900")  # Adjusted size for better layout

        # --- Constants and State Variables ---
        self.square_size = 80
        self.board_size = 8 * self.square_size
        self.colors = ("#f0d9b5", "#b58863")  # Light, Dark
        self.start_fen = 'rnbqkbnr/pppppppp/8/8/8/8/PPPPPPPP/RNBQKBNR w KQkq - 0 1'
        self.piece_map = {
            'p': '♟', 'r': '♜', 'n': '♞', 'b': '♝', 'q': '♛', 'k': '♚',
            'P': '♙', 'R': '♖', 'N': '♘', 'B': '♗', 'Q': '♕', 'K': '♔'
        }
        self.drag_data = {"item": None, "x": 0, "y": 0}

        # --- UI Setup ---
        self.main_frame = tk.Frame(root, bg="#f0f0f0")
        self.main_frame.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)

        # --- FEN Input and Display ---
        self.fen_frame = tk.Frame(self.main_frame, bg="#ffffff", relief=tk.RIDGE, borderwidth=2)
        self.fen_frame.pack(pady=10, padx=10, fill=tk.X)

        self.fen_input_var = tk.StringVar(value=self.start_fen)
        self.fen_input = tk.Entry(self.fen_frame, textvariable=self.fen_input_var, font=("Courier", 12))
        self.fen_input.pack(side=tk.LEFT, fill=tk.X, expand=True, padx=10, pady=10)

        self.set_fen_button = tk.Button(self.fen_frame, text="Apply FEN", command=self.apply_fen, font=("Arial", 10, "bold"))
        self.set_fen_button.pack(side=tk.RIGHT, padx=10, pady=10)
        
        self.current_fen_label = tk.Label(self.main_frame, text="Current Board FEN:", font=("Arial", 10))
        self.current_fen_label.pack(anchor="w", padx=10)
        
        self.current_fen_var = tk.StringVar(value=self.start_fen)
        self.current_fen_display = tk.Entry(self.main_frame, textvariable=self.current_fen_var, state='readonly', font=("Courier", 12), relief=tk.FLAT, bg="#f0f0f0")
        self.current_fen_display.pack(fill=tk.X, padx=10, pady=(0, 10))


        # --- Chessboard Canvas ---
        self.canvas = tk.Canvas(self.main_frame, width=self.board_size, height=self.board_size)
        self.canvas.pack(pady=20)
        
        self.piece_font = font.Font(family='Arial', size=40)
        
        # Draw the initial board
        self.draw_board()
        
        # Set up drag and drop bindings
        self.canvas.tag_bind("piece", "<ButtonPress-1>", self.on_piece_press)
        self.canvas.tag_bind("piece", "<B1-Motion>", self.on_piece_drag)
        self.canvas.tag_bind("piece", "<ButtonRelease-1>", self.on_piece_drop)
        
        # Initialize with starting position
        self.render_board_from_fen(self.start_fen)
        self.render_board_from_fen(self.start_fen)

        # --- Drag and Drop Bindings ---
        self.canvas.tag_bind("piece", "<ButtonPress-1>", self.on_piece_press)
        self.canvas.tag_bind("piece", "<B1-Motion>", self.on_piece_drag)
        self.canvas.tag_bind("piece", "<ButtonRelease-1>", self.on_piece_drop)

    def draw_board(self):
        """Draws the 64 squares of the chessboard."""
        for row in range(8):
            for col in range(8):
                x1 = col * self.square_size
                y1 = row * self.square_size
                x2 = x1 + self.square_size
                y2 = y1 + self.square_size
                color = self.colors[(row + col) % 2]
                self.canvas.create_rectangle(x1, y1, x2, y2, fill=color, outline="")
                self.canvas.create_rectangle(x1, y1, x2, y2, fill=color, tags="square")

    def render_board_from_fen(self, fen):
        """Renders the board state from a FEN string."""
        print(f"Rendering board from FEN: {fen}")
        self.canvas.delete("piece") # Clear existing pieces
        
        try:
            piece_placement = fen.split(' ')[0]
            ranks = piece_placement.split('/')
            
            for rank_idx, rank_str in enumerate(ranks):
                file_idx = 0
                for char in rank_str:
                    if char.isdigit():
                        file_idx += int(char)
                    else:
                        x = file_idx * self.square_size + self.square_size / 2
                        y = rank_idx * self.square_size + self.square_size / 2
                        
                        piece_char = self.piece_map.get(char, '?')
                        
                        self.canvas.create_text(x, y, text=piece_char, font=self.piece_font,
                                                tags=("piece", char))
                        file_idx += 1
            self.update_fen_display()
        except Exception as e:
            print(f"Error parsing FEN: {e}")
            # Optionally show an error to the user
            tk.messagebox.showerror("FEN Error", "Invalid FEN string format.")


    def generate_fen_from_board(self):
        """Generates a FEN string from the current board state."""
        fen = ''
        for rank in range(8):
            empty_squares = 0
            for file in range(8):
                x = file * self.square_size + self.square_size / 2
                y = rank * self.square_size + self.square_size / 2
                
                items = self.canvas.find_overlapping(x-1, y-1, x+1, y+1)
                piece_item = next((item for item in items if "piece" in self.canvas.gettags(item)), None)

                if piece_item:
                    if empty_squares > 0:
                        fen += str(empty_squares)
                        empty_squares = 0
                    
                    tags = self.canvas.gettags(piece_item)
                    piece_char = tags[1] # The second tag is the FEN character
                    fen += piece_char
                else:
                    empty_squares += 1
            
            if empty_squares > 0:
                fen += str(empty_squares)
            
            if rank < 7:
                fen += '/'
        
        # Append static game state for simplicity
        return fen + ' w KQkq - 0 1'

    def update_fen_display(self):
        """Updates the FEN display with the current board state."""
        new_fen = self.generate_fen_from_board()
        self.current_fen_var.set(new_fen)
        self.fen_input_var.set(new_fen)

    def apply_fen(self):
        """Applies the FEN string from the input box to the board."""
        try:
            fen = self.fen_input_var.get().strip()
            print(f"Applying FEN: {fen}")
            if fen:
                self.render_board_from_fen(fen)
            else:
                print("Warning: Empty FEN string")
        except Exception as e:
            print(f"Error in apply_fen: {e}")
            messagebox.showerror("Error", f"Failed to apply FEN: {e}")
            
    def update_board_with_fen(self, fen):
        """Updates the board with a new FEN string from an external source"""
        try:
            print(f"Updating board with FEN: {fen}")
            self.fen_input_var.set(fen)
            self.render_board_from_fen(fen)
            self.root.update()  # Force an immediate update
        except Exception as e:
            print(f"Error updating board: {e}")
            messagebox.showerror("Error", f"Failed to update board: {e}")

    # --- Drag and Drop Event Handlers ---

    def on_piece_press(self, event):
        """Called when a piece is clicked."""
        self.drag_data["item"] = self.canvas.find_closest(event.x, event.y)[0]
        self.drag_data["x"] = event.x
        self.drag_data["y"] = event.y
        self.canvas.lift(self.drag_data["item"]) # Bring piece to the top

    def on_piece_drag(self, event):
        """Called when a piece is dragged."""
        dx = event.x - self.drag_data["x"]
        dy = event.y - self.drag_data["y"]
        self.canvas.move(self.drag_data["item"], dx, dy)
        self.drag_data["x"] = event.x
        self.drag_data["y"] = event.y

    def on_piece_drop(self, event):
        """Called when a piece is dropped."""
        item = self.drag_data["item"]
        if item:
            # Snap to the center of the target square
            col = int(event.x // self.square_size)
            row = int(event.y // self.square_size)
            
            # Prevent dropping outside the board
            col = max(0, min(7, col))
            row = max(0, min(7, row))

            new_x = col * self.square_size + self.square_size / 2
            new_y = row * self.square_size + self.square_size / 2

            # Check if a piece is already on the target square (capture)
            overlapping = self.canvas.find_overlapping(new_x-1, new_y-1, new_x+1, new_y+1)
            for other_item in overlapping:
                if other_item != item and "piece" in self.canvas.gettags(other_item):
                    self.canvas.delete(other_item)

            self.canvas.coords(item, new_x, new_y)
            
            self.drag_data["item"] = None
            self.update_fen_display()


# Global variables to maintain single instance
_root = None
_app = None
_fen_queue = []
_is_running = False

def get_or_create_board():
    """Get existing board instance or create a new one"""
    global _root, _app
    if _root is None:
        _root = tk.Tk()
        _app = ChessboardApp(_root)
        _app.draw_board()  # Initialize the board
        _app.render_board_from_fen(_app.start_fen)  # Set initial position
    return _app

def process_fen_updates():
    """Process any pending FEN updates"""
    global _fen_queue, _root, _app, _is_running
    try:
        if _fen_queue and _app:
            while _fen_queue:
                fen = _fen_queue.pop(0)
                print(f"Processing FEN update: {fen}")
                if _app and _app.root:
                    _app.root.after_idle(lambda: _app.update_board_with_fen(fen))
    except Exception as e:
        print(f"Error processing FEN updates: {e}")
    finally:
        if _is_running and _root:
            _root.after(100, process_fen_updates)

def update_fen(fen_string):
    """Queue a FEN string update"""
    global _fen_queue, _app
    print(f"Queueing FEN update: {fen_string}")
    _fen_queue.append(fen_string)
    if _app and _app.root:
        _app.root.event_generate('<<ProcessFEN>>')

def start_board():
    """Start the board with the initial position"""
    global _is_running, _root, _app
    try:
        _is_running = True
        app = get_or_create_board()
        app.root.bind('<<ProcessFEN>>', lambda e: process_fen_updates())
        app.root.after(100, process_fen_updates)
        app.root.mainloop()
    except Exception as e:
        print(f"Error in start_board: {e}")
    finally:
        _is_running = False

if __name__ == "__main__":
    import sys
    app = get_or_create_board()
    
    # If a FEN string is provided as a command-line argument, use it
    if len(sys.argv) > 1:
        fen_string = sys.argv[1]
        update_fen(fen_string)
    
    start_board()
```
